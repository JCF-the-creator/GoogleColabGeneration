```python
def ConverterInputIdade(idade):
    try:
        idade = int(idade)
        return idade
    except ValueError:
        # Trata o erro de conversão
        print(f"Erro: Idade deve ser um número do tipo inteiro, sem decimais ou caracteres ")
        return None
```

```python
nome = input("Digite seu nome:\n")
idade = None

while idade == None:
  idade = ConverterInputIdade(input("Digite sua idade:\n"))


Cidade = input("Digite sua Cidade:\n")

```

```python
print(f"\nOlá {nome}, você tem {idade} anos e mora na cidade de {Cidade}")
```
