```python
#1 - criar um loop (while) que pergunte ao usuario se ele quer adicionar uma fruta na prateleira
#2 - a prateleira só pode ter 7 produtos
#3 - Retirar o item da lista
#4 - sempre que eu adicionar um produto, mostrar todos seus valores
```

```python
def verificarIndice(valorIndice):
  try:
    valorIndice = int(valorIndice)
    return valorIndice
  except ValueError:
    print(f"Erro: o Índice deve ser um número, sem texto ou caracteres")
    return None
```

```python
def verificarMenu(menu):
  try:
    menu = int(menu)
    return menu
  except ValueError:
      print(f"Erro: o menu deve ser um número, sem texto ou caracteres")
      return None
```

```python
print(f"\nDesafio da prateleira")
prateleira = ['Batata','agua']


print(f"="*60)
print(f"\nPratileira: ")
for indice, item in enumerate(prateleira):
          print(f"\nIndice {indice} Item: {item}")
print(f"="*60)

limite = len(prateleira)

while True:
  menu = None
  acesso = 0
  while menu == None:
    print(f"Digite o que deseja acessar no menu:\n")
    menu = verificarMenu(input(f"\n[1]Adicionar Itens a Pratileira \n[2]Remover algum item \n[3]Fechar Programa\n"))
    if menu in [1, 2, 3]:
      acesso = int(menu)
    else:
      print(f"Digite um valor acessivel no Menu")
      print(f"\n")
      menu = None

  if(acesso == 1):
    while limite < 7:
      prateleira.append(input(f"\nDigite o nome do item a adicionar"))

      print(f"\n")
      print(f"\nItem adicionado! Sua pratileira possui um: ")
      print(f"\n{prateleira}, Prateleira com um total de {len(prateleira)} item")
      print(f"\n")
      limite = len(prateleira)

      continuar = None
      while continuar == None:
        continuar = input(f"\nDeseja continuar? S/N").lower()
        if(continuar == "n" or continuar == "não" or continuar == "nao"):
          continuar = False
        elif(continuar == "s" or continuar == "sim"):
          continuar = True
        else:
          print(f"\n Digite sim ou não")
          continuar = None

      if(continuar == False):
        break

    if(limite == 7):
      print(f"\nPrateleira cheia, maximo de 7 itens")
      print(f"\n{prateleira}")


  if(acesso == 2):
    remover = input(f"\nDeseja remover algum item a pratileira? S/N").lower()
    if(remover == "s" or remover == "sim"):
      limiteRemover = len(prateleira)
      while limiteRemover > 0:
        limiteRemover = len(prateleira)
        print(f"\n")
        print(f"\nPratileira: ")
        for indice, item in enumerate(prateleira):
          print(f"\nIndice {indice} Item: {item}")
        print(f"\n")

        removerIndice = 0
        while removerIndice == 0:
          valorIndice = None
          while valorIndice == None:
            valorIndice = verificarIndice(input(f"Digite o índice exato do produto que deseja remover:\n"))

          try:
            prateleira.pop(valorIndice)
            print(f"\n")
            for indice, item in enumerate(prateleira):
              print(f"\nIndice {indice} Item: {item}")
            print(f"\n")
            removerIndice = 1
          except IndexError:
            print(f"\nDigite um valor de índice válido")
            print(f"\n")
            removerIndice = 0

        limiteRemover = len(prateleira)
        print(f"="*60)
        print(f"\n")
        print(f"\nNova Prateleira: ")
        for indice, item in enumerate(prateleira):
          print(f"\nIndice {indice} Item: {item}")
        print(f"\n")
        print(f"="*60)

        continuar = None
        while continuar == None:
          continuar = input(f"\nDeseja continuar? S/N").lower()
          if(continuar == "n" or continuar == "não" or continuar == "nao"):
            continuar = False
          elif(continuar == "s" or continuar == "sim"):
            continuar = True
          else:
            print(f"\n Digite sim ou não")
            continuar = None

        if(continuar == False):
          break

      if(limiteRemover == 0):
        print(f"\nPratileira Vazia")

  if(acesso == 3):
    fechar = None
    while fechar == None:
      fechar = input(f"\nDeseja fechar o programa? S/N").lower()
      if(fechar == "n" or fechar == "não" or fechar == "nao"):
        fechar = True
      elif(fechar == "s" or fechar == "sim"):
        fechar = False
      else:
        print(f"\n Digite sim ou não")
        fechar = None

    if(fechar == False):
      print(f"\nFinalizado o Programa")
      break
```
