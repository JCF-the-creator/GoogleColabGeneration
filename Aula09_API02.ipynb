```python
#usando outra API

import requests

def buscar_clima():

  #nunca informe a API key no programa, isso deixa uma brecha grande de segurança

  api_key = "" #Adicionar a API key do site

  url = "https://api.openweathermap.org/data/2.5/weather"

  latitude = float(input('informe a latitude desejada'))
  longitude = float(input('informe a longitude desejada'))

  parametros = {
      'appid' : api_key,
      'lang': "pt_br",
      'lat': latitude,
      'lon': longitude,
      'units': 'metric'
  }

  # fazer as requisições
  resposta = requests.get(url, params=parametros)

  # verificar se esta correto
  if resposta.status_code == 200:
    # transformar em json
    dados = resposta.json()

    # vamos coletar as informações
    temp = dados['main']['temp']
    sensacao =  dados['main']['feels_like']

    print(f'temperatura {temp}')
    print(f'sensação {sensacao}')
  else:
    print('erro ao exibir a API')

buscar_clima()
```
