```python
# o nosso desafio é mostrar os produtos de um mercado
# quero poder escolher os produtos, para ver infos
# quero poder adicionar novos produtos
# quero poder excluir um produto do mercado

#----


class Produto:
  def __init__(self, nome, preco):
    self.nome = nome
    self.preco = preco
```

```python

listaDeProdutos = []

listaDeProdutos.append(Produto("Detergente ype", 2.50))
listaDeProdutos.append(Produto("Refrigerante Gengibirra", 5))
listaDeProdutos.append(Produto("Arroz", 7.25))
listaDeProdutos.append(Produto("Feijão", 4))
listaDeProdutos.append(Produto("Água mineral", 3.45))
```

```python
def verProdutos():
  indice = 0
  for produto in listaDeProdutos:
      print(f"Índice {indice}: {produto.nome}, Preço: R$ {produto.preco:.2f}")
      indice = indice + 1
```

```python
def verificarValor(valor):
  try:
    valor = int(valor)
    return valor
  except ValueError:
    print(f"Erro: o Preço deve ser um número, sem texto ou caracteres")
    return None
```

```python
def verificarMenu(menu):
  try:
    menu = int(menu)
    return menu
  except ValueError:
      print(f"Erro: o menu deve ser um número, sem texto ou caracteres")
      return None
```

```python
def verificarIndice(valorIndice):
  try:
    valorIndice = int(valorIndice)
    return valorIndice
  except ValueError:
    print(f"Erro: o Índice deve ser um número, sem texto ou caracteres")
    return None
```

```python
def verMenu():
  menu = None
  acesso = 0
  while menu == None:
    print(f"Digite o que deseja acessar no menu:\n")
    menu = verificarMenu(input(f"\n[1]Ver lista de Produtos [2] Adicionar item na Lista [3] Excluir itens na Lista\n"))
    if menu in [1, 2, 3]:
      acesso = int(menu)
    else:
      print(f"Digite um valor acessivel no Menu")
      print(f"\n")
      menu = None


  if (acesso == 1):
    print(f"\n")
    verProdutos()
    print(f"\n")

  if (acesso == 2):
    print(f"\n")
    nomeProduto = input(f"Digite o nome do produto:\n")
    valor = None
    while valor == None:
      valor = verificarValor(input(f"Digite o preço do Produto: \n"))

    listaDeProdutos.append(Produto(nomeProduto, valor))
    print(f"\n")
    verProdutos()
    print(f"\n")

  if(acesso == 3):
    print(f"\n")

    removerIndice = 0
    while removerIndice == 0:
      valorIndice = None
      while valorIndice == None:
        valorIndice = verificarIndice(input(f"Digite o índice exato do produto que deseja remover:\n"))

      try:
        listaDeProdutos.pop(valorIndice)
        print(f"\n")
        verProdutos()
        print(f"\n")
        removerIndice = 1
        return removerIndice
      except IndexError:
        print(f"\nDigite um valor de índice válido")
        print(f"\n")
        removerIndice = 0
        return removerIndice


    print(f"\n")
    verProdutos()
    print(f"\n")
```

```python
while True:
  verMenu()



```
