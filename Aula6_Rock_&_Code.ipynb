```python
#- Desenvolver pipeline no Google Colab usando Pandas para leitura e validação dos 3 arquivos. x

#- Executar a concatenação vertical garantindo alinhamento de colunas. x

#- Tratamento de dados: Criar coluna Faturamento_Total e converter datas.

#- Exportar base consolidada final: trimestre_completo.csv.

import pandas as pd
```

```python
df_jan = pd.read_csv("/content/vendas_janeiro.csv")
display(df_jan.head())
```

```python
df_fev = pd.read_csv("/content/vendas_fevereiro.csv")
display(df_fev.head())
```

```python
df_mar = pd.read_csv("/content/vendas_marco.csv")
display(df_mar.head())
```

```python
df_dados_juntos = pd.concat([df_jan, df_fev, df_mar], ignore_index=True)
display(df_dados_juntos)
```

```python
df_dados_juntos['Faturamento'] = df_dados_juntos['Preço'] * df_dados_juntos["Qtd"]
df_dados_juntos.to_csv("trimestre_completo.csv", index=False)
display(df_dados_juntos)
```
