```python
# vamos criar nossa base de dados vazia
import getpass as gp

usuarios = {}

# função de cadastro e de login
def cadastrar():
  novo_usuario = input('informe o nome do usuario')
  nova_senha = gp.getpass('informe a senha')

  # para adicionar um valor no dicionario
  # dicionario[chave] = valor
  usuarios[novo_usuario] = nova_senha
  print('usuario cadastrado com sucesso')


def login():
  usuario=input("informe seu usuario")
  senha = gp.getpass('informe sua senha')
  #vamos verificar se o usuario digitado esta no dicionario usuarios
  # e senha informado é o valor da chave
  if usuario in usuarios and  usuarios[usuario] ==senha :
    print('login reaizado com sucesso')

  else:
      print('usuario e /ou senha incorretos')

#executa o sistema
while True:
  print(f"\n")
  print(f"\n1 - Cadastrar novo usuário")
  print(f"\n2 - ver usuários cadastrados")
  print(f"\n3 - ver senhas cadastradas")
  print(f"\n4 - ver usuário e senhas")
  print(f"\n5 - ver usuários e senhas")
  print(f"\n0 - sair do sistema")

  op = input(f"\nInforme sua escolha:\n")
  print(f"\n")

  if(op == "1"):
    cadastrar()
    print(f"\n")
  elif(op == "2"):
    print(f"{usuarios.keys()}")
    print(f"\n")
  elif(op == "3"):
    print(f"{usuarios.values()}")
    print(f"\n")
  elif(op == "4"):
    print(f"{usuarios.items()}")
    print(f"\n")
  elif(op == "5"):
    login()
    print(f"\n")
  elif(op == "0"):
    print(f"\nSistema Finalizado")
    break
  else:
    print("ERRO, Digite um valor válido do menu, de 0 a 5")
    print(f"\n")
```
